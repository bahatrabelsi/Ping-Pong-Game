# **ML Project**


# **Load DATASET**


```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/dataprofessor/data/refs/heads/master/delaney_solubility_with_descriptors.csv")
df
```

## Data Seperation : X & Y

```python
y = df["logS"]
y
```

```python
x = df.drop("logS" , axis = 1)
x
```

## Data splitting

```python
from sklearn.model_selection import train_test_split

x_train , x_test , y_train , y_test = train_test_split(x , y , test_size = 0.2 , random_state = 100)
x_test
```

```python
x_train
```

# **Model Building**

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train , y_train)
```

### Model Predection

```python
y_lr_train_pred = lr.predict(x_train)
y_lr_test_pred = lr.predict(x_test)
print(y_lr_train_pred)
```

```python
print(y_lr_test_pred)
```

### Evaluate Model Performance

```python
from sklearn.metrics import mean_squared_error , r2_score

lr_train_mse = mean_squared_error(y_train , y_lr_train_pred)
lr_train_r2 = r2_score(y_train , y_lr_train_pred)

lr_test_mse = mean_squared_error(y_test , y_lr_test_pred)
lr_test_r2 = r2_score(y_test , y_lr_test_pred)
```

```python
print("MSE Train : ",lr_train_mse )
print("R2 Train : ",lr_train_r2 )
print("MSE Test : ",lr_test_mse )
print("R2 Test : ",lr_test_r2 )
```

```python
lr_result = pd.DataFrame(["Linear Regression" , lr_train_mse , lr_train_r2 , lr_test_mse , lr_test_r2]).transpose()
lr_result.columns = ["Model" , "Training MSE" , "Training R2" , "Test MSE" , "Test R2"]
lr_result
```

## **Random Forest**

### Training the Model

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(max_depth = 2 , random_state = 100)
rf.fit(x_train , y_train)
```

### Model Predection

```python
y_rf_train_pred = rf.predict(x_train)
y_rf_test_pred = rf.predict(x_test)
print(y_rf_train_pred)
```

### Evaluate Model Performance

```python
from sklearn.metrics import mean_squared_error , r2_score

rf_train_mse = mean_squared_error(y_train , y_rf_train_pred)
rf_train_r2 = r2_score(y_train , y_rf_train_pred)

rf_test_mse = mean_squared_error(y_test , y_rf_test_pred)
rf_test_r2 = r2_score(y_test , y_rf_test_pred)
```

```python
rf_result = pd.DataFrame(["Random Forest" , rf_train_mse , rf_train_r2 , rf_test_mse , rf_test_r2]).transpose()
rf_result.columns = ["Model" , "Training MSE" , "Training R2" , "Test MSE" , "Test R2"]
rf_result
```

## **Model Comparison**

```python
df_models = pd.concat([lr_result , rf_result] , axis = 0)
df_models
```

# **Data Visualisation of Prediction**
